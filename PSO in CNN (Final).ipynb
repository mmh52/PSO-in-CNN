```python
import numpy as np
import random
from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target
Y=y.transpose()
```

```python
n_input=len(X[0]);n_sample=len(X)
n_hidden=20;n_output=3;n_particle=20
```

```python
np.random.seed(110)
v=np.random.rand(163)
W1=np.reshape(v[0:80],(len(X[0]),n_hidden))
b1=np.reshape(v[80:100],(n_hidden,))
W2=np.reshape(v[100:160],(n_hidden,n_output))
b2=np.reshape(v[160:163],(n_output,))
# print(W1,W2,b1,b2)
```

```python
def soft_max(Y):
    for i in range(len(Y)):
        Y[i]=np.exp(Y[i])
        Y[i]=Y[i]/sum(Y[i])
    return Y
```

```python
def Model(X,W1,W2,b1,b2):
    y1=X.dot(W1)+b1;
    y1=np.tanh(y1);
    y_hat=y1.dot(W2)+b2
    return y_hat

Model(X,W1,W2,b1,b2)
```

```python
def Loss_linear_vector(v):
#     W1=np.reshape(v[0:80],(len(X[0]),n_hidden))
#     w2=np.reshape(v[100:160],(n_hidden,n_output))
    y_hat=Model(X,W1,W2,b1,b2)
    y_hat=soft_max(y_hat)
    
    L=np.zeros(n_sample)
    for i in range(n_sample):
        L[i]=-np.log(y_hat[i][Y[i]])
    L=sum(L)/len(L);
    
    return L
    
    
print(Loss_linear_vector(v))
```

```python
def Loss_linear(x):
    n_particle = x.shape[0]
    loss=[Loss_linear_vector(x[i]) for i in range(n_particle)]
    return np.array(loss)
```

```python
x=np.random.rand(n_particle,163)
print(Loss_linear(x))
```

```python
c1=0.6;c2=0.2
w0=0.9
r1=random.uniform(0, 1)
r2=random.uniform(0, 1)
vel=np.random.rand(n_particle,163)*0.01
p_best=P;
       
### Set g_best
L=Loss_linear(X,y,W1,W2,b1,b2,P)
g_best=P[np.argmin(L)]
# print(g_best)


iterations=100
for j in range(iterations):
    g_best=np.repeat(np.reshape(g_best,(1,-1)),n_particle,axis=0)
    for i in range(n_particle):
        if(Loss_linear_vector(X,y,b1,b2,n_input,n_hidden,n_output,P[i])<Loss_linear_vector(X,y,b1,b2,n_input,n_hidden,n_output,p_best[i])):
                p_best[i]=P[i]
        vel=w0*vel+c1*r1*(p_best-P)+c2*r2*(g_best-P)
        P=P+vel
        L=Loss_linear(X,y,W1,W2,b1,b2,P)
        g_best=P[np.argmin(L)]
print(Loss_linear_vector(X,y,b1,b2,n_input,n_hidden,n_output,g_best))
```
